```python
import emoji
import collections
import pandas as pd

from unidecode import unidecode
```

```python
def extract_emojis(str):
    return ''.join(c for c in str if c in emoji.UNICODE_EMOJI)

def get_emoji_list(str):
    new_str = ""
    for char in str:
        new_str += " {} ".format(char)
    return [v for v in map(lambda x: x.strip(), new_str.split(" ")) if v != ""]
```

```python
def count_emojis(df):
    all_emojis = []
    for index, row in df.iteritems():
        emoji_only = extract_emojis(row)
        if emoji_only is not ''
            emos = get_emoji_list(emoji_only)
            all_emojis = all_emojis + emos
    return collections.Counter(all_emojis)
```
